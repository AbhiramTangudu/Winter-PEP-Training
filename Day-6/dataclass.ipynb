# Decorators

The @dataclass decorator automatically generates the following methods for you:

    1 init(): initializes the object and assigns the provided values to the attributes'
    2 repr(): provides a string representation of the object/
    3 eq(): Implements equality comparision b/w two objs of the class based on their attributes.

```python
from curses.textpad import rectangle
from dataclasses import dataclass
```

```python
@dataclass
class Rectangle:
    width:int
    height:int
    color:str="blue"

rectangle1=Rectangle(10, 20)
rectangle2=Rectangle(15, 25, "red")
```

```python
print(rectangle1)
print(rectangle2)
```

```python
@dataclass(frozen=True)
class Point:
    x:int
    y:int
```

```python
point=Point(3, 4)
```

```python
point.x,point.y
```

### Inheritance

```python
@dataclass
class Person:
    name: str
    age: int
    
```

```python
@dataclass
class Employee(Person):
    employee_id:str
    department:str
```

```python
@dataclass
class Address:
    street:str
    city:str
    zip_code:str
    
```

```python
@dataclass
class Person:
    name:str
    age:str
    address:Address
```

```python
address=Address('123 Main St', 'Bangalore', '12345')
person=Person('Abhi', 20, address)
```

```python
person.address.city
```

```python
@dataclass
class Student:
    name:str
    regno:str
    rollno:str
    
```

```python
@dataclass
class Marks:
    start:int
    mid:int
    end:int

@dataclass
class StudentRecord:
    student:Student 
    marks:Marks
record=StudentRecord(
    Student('Abhi', '12313977', '54'),
    Marks(85, 90, 95)
)

record.marks.end
```

# Inheritance

```python
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
    
    def driving(self):
        print("The car is driving")

class Audi(Car):
    def __init__(self, windows, doors, enginetype, horsepower):
        super().__init__(windows, doors, enginetype)
        self.horsepower=horsepower

    def selfdriving(self):
        print("IT is a self driving car")
audiq7=Audi(4, 5, 'Diesel', 200)

print(audiq7.horsepower)
print(audiq7.windows)
audiq7.driving()
audiq7.selfdriving()
car1=Car(4, 4, 'Diesel')
print(car1)
print(audiq7)

print(dir(audiq7))
print(dir(car1))
```

## Class Methods And Class Variables 

```python
class Car:
    base_price=100000
    def __init__(self, windows, doors, power):
        self.windows=windows
        self.doors=doors
        self.power=power
    
    def what_base_price(self):
        print("the base price of the car is {}".format(self.base_price))
    @classmethod
    def revise_base_price(cls, inflation):
        cls.base_price=cls.base_price + cls.base_price*inflation
```

```python
Car.revise_base_price(0.17)
```

```python
Car.base_price
```

```python
car1=Car(4, 5, 2000)
```

```python
car1.base_price
```
