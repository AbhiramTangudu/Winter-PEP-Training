# Python OOPS-Public , Protected And Private

#### Class Variables are public

```python
class Car():
    def __init__(self, windows, doors, enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
```

```python
audi=Car(4,5,"Diesel")
```

```python
audi.windows=5
```

```python
audi.windows
```

#### Class Variables are Protected

```python
class Car():
    def __init__(self, windows, doors, enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
```

```python
class truck(Car):
    def __init__(self, windows, doors, enginetype, horsepower):
        super().__init__(windows, doors, enginetype)
        self.horsepower=horsepower
```

```python
truck._doors=5
```

```python
truck._doors
```

#### Class Variables are Private

```python
class Car():
    def __init__(self, windows, doors, enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
```

```python
audi=Car(4,4,"Diesel")
```

```python
audi._Car__doors=5
```
